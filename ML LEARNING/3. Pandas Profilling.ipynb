```python
from IPython import get_ipython
from IPython.display import display
# %%
import pandas as pd
# %%
df = pd.read_csv('/content/drive/MyDrive/ai ml sample dataset/exoplanet_dataset.csv')
# %%
df.head()
# %%
!pip install ydata_profiling
# %%
from ydata_profiling import ProfileReport
from ydata_profiling.config import Settings # Import Settings

# %%
df.info()
# %%
# Create a Settings object
settings = Settings()
# Update the settings with your desired configuration
settings.vars.num.chi_squared_threshold = 0

# Now pass the Settings object to ProfileReport
prof = ProfileReport(df, config=settings)
prof.to_file(output_file='exoplanet_dataset.html')

# %%
```
